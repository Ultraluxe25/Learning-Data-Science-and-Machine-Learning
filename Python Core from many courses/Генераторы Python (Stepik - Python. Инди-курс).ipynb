**Генератор** – итератор, элементы которого можно итерировать только один раз

**Итератор** – объект, который поддерживает функцию next(). Помнит о том, какой элемент будет браться следующим.

**Итерируемый объект** – объект, который предоставляет возможность обойти поочередно свои элементы. Может быть преобразован к итератору.

### Генеретар сразу является итератором и может использовать функцию next()


### Генератор можно обойти только один раз!


### Генератор может работать с данными, которые не умещаются в памяти за счет ленивых вычислений

### Переменная с не хранит все эти элементы, а в момент времени получает их по одному. Используется если есть большие данные или большие вычисления

### Найти длину генератора нельзя. Функция len() вызовет исключение. Индексацию тоже нельзя к нему применить.

### Если генератор явно нужно привести к списку, то используйте list(), т.к. [] все равно сохранит внутри пустого списка не коллекцию, а один объект генератор:



### При создании списка из генератора, происходит полный обход по нему и повторный список из него будет уже пустым.


### Вычисляя max(), мы проходимся по всему генератору и снова по нему уже не пройтись

### Генератор запоминает на какой итерации он остановился и продолжает с того места, где он закончил
