```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = 2 * x
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph');
```

```python
# Change limits
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph')

plt.xlim(0, 6)
plt.ylim(0, 9);
# plt.savefig('Graph.png')  # Save graph in png-format
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])  # Build axes
axes.plot(x, y);
```

```python
# First two plots on figure
fig = plt.figure()

# Big axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

# Small axes
axes2 = fig.add_axes([0.15, 0.35, 0.5, 0.5])
axes2.plot(x, y);
```

```python
fig = plt.figure()

# Big axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8_000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Pow in 4')

# Small axes
axes2 = fig.add_axes([0.15, 0.35, 0.5, 0.5])
axes2.plot(a, b)
axes2.set_xlim(4, 5)
axes2.set_ylim(0, 500)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Increased');
```

```python
# Logarithmic scale
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axes[0].plot(x, x ** 2, x, np.exp(x))
axes[0].set_title('Normal scale')

axes[1].plot(x, x ** 2, x, np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Logarithmic scale (y)');
```
