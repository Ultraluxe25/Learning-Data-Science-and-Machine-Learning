### Т.к. производная может быть очень большой, то вводят коэффициент, который уменьшает итоговое значение


### На практике чтобы не попадать в локальный минимум перебирают разные Xo стартовые точки, чтобы определить меньшее значение функции (глобальный минимум)

### Также переберают разные значения шага сходимости, а также меняют его по ходу обучения

```python
import numpy as np


def func(x):
    return 0.5 * x + 0.2 * x ** 2 - 0.1 * x ** 3


def df(x):
    # здесь выражение производной функции f(x)
    return 0.5 + 0.4 * x - 0.3 * x ** 2


coord_x = np.arange(-5.0, 5.0, 0.1) # значения по оси абсцисс
coord_y = func(coord_x) # значения по оси ординат (значения функции)

# здесь продолжайте программу
η = 0.01
x = -4
N = 200

for _ in range(N):
    x = x - η * df(x)
    # print(x)
x
```

```python
import numpy as np

# исходная функция, которую нужно аппроксимировать моделью a(x)
def func(x):
    return 0.1 * x**2 - np.sin(x) + 5.

coord_x = np.arange(-5.0, 5.0, 0.1) # значения по оси абсцисс [-5; 5] с шагом 0.1
coord_y = func(coord_x) # значения функции по оси ординат

sz = len(coord_x)	# количество значений функций (точек)
eta = np.array([0.1, 0.01, 0.001, 0.0001]) # шаг обучения для каждого параметра w0, w1, w2, w3
w = np.array([0., 0., 0., 0.]) # начальные значения параметров модели
N = 200 # число итераций градиентного алгоритма

s = np.array([[1, x, x ** 2, x ** 3] for x in coord_x])
for _ in range(N):
    a_x = s @ w  # y_pred
    dQ_w = 2 * ((a_x - coord_y) * s.T).mean(axis=1)
    w = w - eta * dQ_w

Q = np.square(a_x - coord_y).sum() / sz
Q
```
