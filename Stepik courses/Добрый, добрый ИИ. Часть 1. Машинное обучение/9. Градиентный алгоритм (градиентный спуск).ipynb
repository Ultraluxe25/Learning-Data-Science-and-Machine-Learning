```python
import numpy as np

#  ,     a(x)
def func(x):
    return 0.1 * x**2 - np.sin(x) + 5.

coord_x = np.arange(-5.0, 5.0, 0.1) #     [-5; 5]   0.1
coord_y = func(coord_x) #     

sz = len(coord_x)	#    ()
eta = np.array([0.1, 0.01, 0.001, 0.0001]) #      w0, w1, w2, w3
w = np.array([0., 0., 0., 0.]) #    
N = 200 #    

s = np.array([[1, x, x ** 2, x ** 3] for x in coord_x])
for _ in range(N):
    a_x = s @ w  # y_pred
    dQ_w = 2 * ((a_x - coord_y) * s.T).mean(axis=1)
    w = w - eta * dQ_w

Q = np.square(a_x - coord_y).sum() / sz
Q
```
