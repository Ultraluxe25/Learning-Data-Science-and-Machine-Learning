#  <span style="color: green">Parsing Engineer с нуля до junior

### Парсинг — это процесс анализа данных или текста, при котором выделяются её значимые части, и преобразуются в формат, удобный для дальнейшего использования.

### Скрепинг — это процесс извлечения данных из различных источников (из документов PDF, Excel, Word)

### Веб-скреперы — программы или скрипты, которые отправляют HTTP-запросы к веб-сайту, получают HTML-код и извлекают из него нужную информацию. Часто используются для получения данных, которые сложно скачать стандартными способами (например, через API).

### JSON или XML файл может подходить примером к полуструктурированным данным.
