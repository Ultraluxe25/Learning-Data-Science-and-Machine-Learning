### SVD (сингулярное разложение) позволяет разделить сложную матрицу на более простые части, которые легче анализировать и использовать для различных целей, например, для уменьшения размерности данных, устранения шума или поиска скрытых закономерностей.

```python
import numpy as np

arr = np.eye(10)
arr
```

```python
mat_A = np.random.randint(1, 6, (5, 5))
mat_A
```

```python
means = np.mean(mat_A, axis=1)
means
```
