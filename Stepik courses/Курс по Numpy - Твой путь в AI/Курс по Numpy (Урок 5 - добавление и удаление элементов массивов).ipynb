
### Оказывается можно создавать массив, не заполняя его значениями!)

### Количество элементов могут отличать только те оси тензора, по которым добавляются элементы!


### Второй параметр метода отвечает за индексы элементов тензора, которые нужно удалить


### Случайным образом добавляется 1000 картинок 28 на 28 пикселей. Работает медленно, т.к. np.append() 1000 раз создает копию массива и каждый раз выделяет много памяти


### Данный метод отработал в 200 раз быстрее за счет расширения исходного массива и залития пустых 1000 изображений нулями

```python
%%time
import numpy as np

arr = np.empty(shape=[1000, 28, 28])
new_arr = np.resize(a=arr, new_shape=[2000, 28, 28])

for i in range(1000):
    new_arr[1000 + i] = np.random.randint(low=0, high=255, size=28*28).reshape([28, 28])
new_arr
```
