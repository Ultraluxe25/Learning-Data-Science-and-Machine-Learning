### Primary key (PK)  - первичный ключ. Это отдельное поле (или комбинация полей), которые однозначно определяют запись в течение всего срока жизни таблицы. Ни одно из полей, являющихся частью первичного ключа, не может содержать значение NULL (не может быть пустым). Таблица может иметь только один первичный ключ.
