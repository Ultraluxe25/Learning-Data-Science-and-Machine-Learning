### Airflow широко признан как инструмент для управления процессами ETL (Extract, Transform, Load).

Airflow представляет собой как сервис с пользовательским интерфейсом, так и библиотеку на Python.

В продакшене часто используется развертка Airflow не просто в Docker как у нас, а в Kubernetes-кластере, для закрытия потребностей в отказоустойчивости и масштабируемости. При развертке в кубере можно дополнительно обеспечить гарантию доставки сообщений с помощью различных брокеров сообщений.

SequentialExecutor совсем простой парень - он умеет запускать задания только последовательно. LocalExecutor уже похитрее, он умеет запускать задания параллельно.
