```python
# Load Iris data
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target
```

# Logistic Regression

```python
# Use logistic regression
import warnings
from sklearn.linear_model import LogisticRegression

warnings.filterwarnings("ignore")

logreg = LogisticRegression()
logreg.fit(X, y)
logreg.predict(X)
```

```python
# Count predictions
y_pred = logreg.predict(X)
len(y_pred)
```

```python
from sklearn import metrics
metrics.accuracy_score(y, y_pred)
```

# K-nearest neighbors

```python
# KNN (k = 5)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
y_pred = knn.predict(X)
metrics.accuracy_score(y, y_pred)
```

```python
# KNN (k = 1)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
y_pred = knn.predict(X)
metrics.accuracy_score(y, y_pred)
```

### 1.0 doesn't mean that knn (k = 1) is a

# Low number of n_neighbors increase probability of overfitting

```python
# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=26)
```

```python
# Logistic regression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# Again knn=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
# Again knn=1
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```

## Here better use KNN with 5 neighbors

```python
# Elbow method to find best n_neighbors value
import matplotlib.pyplot as plt

scores = []
for k in range(1, 26):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    
plt.plot(range(1, 26), scores)
plt.grid(visible=True)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy');
```

## Best accuracy will be with KNN (k=10)
